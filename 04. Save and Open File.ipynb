# Save and Open File

```python
import pandas as pd
data = {
    'name' : ['Sally', 'David', 'Daniel', 'Peter'],
    'school' : ['Biola', 'UCI', 'Navy', 'Emery'],
    'height' : [157, 173, 175, 130],
    'English' : [97, 78, 95, 100] ,
    'Math' : [89, 100, 78, 97],
    'Science' : [88, 98, 94, 97],
    'SWskills' : ['C++', 'Python', 'Java', 'Swift']
}
df = pd.DataFrame(data, index = ['1st', '2nd', '3rd', '4th'])
df.index.name = 'Sib. #'
df
```

## Save
### To csv file

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False) # save by removing the index
```

### To txt file

```python
df.to_csv('score.txt', sep='\t')
```

### To excel file

```python
df.to_excel('score.xlsx')
```

***
## Open
### open csv file

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows = 1) # skip number of rows
df
```

```python
df = pd.read_csv('score.csv', skiprows = [1, 3])
df
```

```python
df = pd.read_csv('score.csv', nrows = 2) # bring only first 2
df
```

```python
df = pd.read_csv('score.csv', skiprows = 1, nrows = 2) # skipp the first row and bring 2 from there
df
```

### open txt file

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='Sib. #')
df
```

### open excel file

```python
df = pd.read_excel('score.xlsx', index_col='Sib. #')
df
```
