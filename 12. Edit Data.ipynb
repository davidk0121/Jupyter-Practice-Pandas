# 12. Edit Data

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='Sib. #')
df
```

## Edit Column

```python
df['school'].replace({'Emery':'Betty'})
```

```python
df['school'].replace({'Emery':'Betty'}, inplace = True)
df
```

```python
df['school'] = df['school'].str.lower()
df
```

```python
df['school'] = df['school'] + ' school'
df
```

## Add Column

```python
df['total'] = df['English'] + df['Math'] + df['Science']
df
```

```python
df['result'] = 'Fail'
df
```

```python
df.loc[df['total'] > 270, 'result'] = 'Pass'
df
```

## Delete Column

```python
df.drop(columns = ['result'])
```
