# 11. Data Sorting

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='Sib. #')
df
```

```python
df.sort_values('height') # sort in ascending order
```

```python
df.sort_values('height', ascending = False) # sort in Descending order
```

```python
df.sort_values(['height', 'English'], ascending = False)
```

```python
df.sort_values(['height', 'English'], ascending = [False, True])
```

```python
df.sort_index(ascending = False)
```
