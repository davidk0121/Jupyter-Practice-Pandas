## 3. Index

```python
import pandas as pd
data = {
    'name' : ['Sally', 'David', 'Daniel', 'Peter'],
    'school' : ['Biola', 'UCI', 'Navy', 'Emery'],
    'height' : [157, 173, 175, 130],
    'English' : [97, 78, 95, 100] ,
    'Math' : [89, 100, 78, 97],
    'Science' : [88, 98, 94, 97],
    'SWskills' : ['C++', 'Python', 'Java', 'Swift']
}
df = pd.DataFrame(data, index = ['1st', '2nd', '3rd', '4th'])
df
```

```python
df.index
```

## Set Index name

```python
df.index.name = 'Sib. #'
df
```

# Reset Index

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # change df temporary
```

```python
df.reset_index(drop=True, inplace=True) # change df 
df
```

## Index setting

```python
df.set_index('name') # change temporary
```

```python
df.set_index('name', inplace=True)
df
```

## Sorting Index

```python
df.sort_index() # ascending order
```

```python
df.sort_index(ascending = False) # descending order
```
