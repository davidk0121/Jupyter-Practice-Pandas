# 13. Apply Function

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='Sib. #')
df
```

## Apply function in Data

```python
def add_cm(height):
    return str(height) + 'cm'

df['height'] = df['height'].apply(add_cm)
df
```

```python
def capitalize(lang):    # same as df['SWskills'].str.capitalize()
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['SWskills'] = df['SWskills'].apply(capitalize)
df 
```
